```python
import pandas as pd

pd.set_option("display.max_columns", None)

df = pd.read_csv(r'C:\Users\imffb\project\EDA\starting_point.csv')
df = df.drop(columns=['Unnamed: 0'])
df.head(50)
```

```python
df['Sqft'] = df['Sqft'].str.split('-').str[1]
df['Yr_built'] = df['Yr_built'].str.split('-').str[1]
print(len(df[df['Sqft'].isnull()]))
print(len(df[df['Yr_built'].isnull()]))
```

```python
df.dropna(subset=['Sqft','Yr_built'], inplace=True)
```

```python
all_sqft = sorted(list(set(df['Sqft'].astype(int))))

df = df.reset_index()
for x in range(0,len(df)):
    df.loc[x,'Sqft'] = all_sqft.index(int(df.loc[x,'Sqft']))
df.tail(50)
```

Approx Square Footage and Approx Age columns were ranges, so I changed them to their indexes in relation to all square footages, to test for relevancy. Null values are removed.

```python
df = df.reindex(sorted(df.columns), axis=1)
df.head(50)
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np

plt.figure(figsize=(10,10))
data = df.groupby('S_r')['S_r'].count()
_, _, autotexts = plt.pie(data, startangle=90, colors=['#53a0fc','#10263d'], autopct='%1.1f%%')
autotexts[0].set_color('black')
autotexts[1].set_color('white')
plt.legend(labels = ['Lease','Sale'], loc='lower center')
plt.title('Condos for Sale vs Condos for Lease', fontsize='20', y=0.93)
plt.show()
```

Condos for sale vs condos for lease. Lease and sale prices vary greatly.

```python
for rooms in sorted(list(df_toronto_sale['Bath_tot'].unique())):
    
    df_toronto_sale_rooms = df_toronto_sale[df_toronto_sale['Bath_tot']==rooms].copy()
    if len(df_toronto_sale_rooms) < 10:
        continue
    date_min = list((df_toronto_sale_rooms['Input_year_month']).unique())[0]
    date_max = list((df_toronto_sale_rooms['Input_year_month']).unique())[-1]
    dates = []
    for year in range(int(date_min.split('-')[0]), int(date_max.split('-')[0])+1):
        if year == int(date_min.split('-')[0]):
            x = int(date_min.split('-')[1])
        else:
            x = 1
        while x < 13:
            dates.append('{}-{}'.format(str(year),str(x)))
            if year == int(date_max.split('-')[0]) and x == int(date_max.split('-')[1]):
                break
            x+=1

    means = []
    for x in dates:
        try:
            means.append(statistics.mean(df_toronto_sale_rooms.loc[df_toronto_sale_rooms['Input_year_month'] == x, 'Orig_dol']))
        except:
            means.append(0)
    
    #Get rid of outliers (3 standard deviations away from the mean)
    klist = np.array(means)
    avglatlist = np.array(dates)
    indices = np.where(np.absolute(zscore(klist)) > 3)[0]
    for outlier in indices:
        means[outlier] = 0

    fig = plt.figure(figsize = (10,10))
    plt.bar(dates, means)
    plt.xticks(rotation = 90)
    plt.xlabel('Month and Year')
    plt.ylabel('Mean Price')
    plt.title('Mean Sale Price Evolution by Time for Toronto; {} Bathrooms'.format(rooms))
    plt.show
```
